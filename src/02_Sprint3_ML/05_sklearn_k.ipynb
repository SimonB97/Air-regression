```python
import pandas as pd
import functions as func
import orga_functions as org
import json
```

```python
df = pd.read_csv(org.path("01_AirQuality_processed.csv"), sep=';', index_col='date')
df
```

### Define target & feature list

```python
target = 'ah_target'
```

```python
# define features list
#features = ['co_gt', 'pt08_s1_co', "ah"]
features = ['co_gt', 'pt08_s1_co']
```

### Shift

```python
# shift abs humidty values by 6 hrs
target = 'ah_target'
df[target] = df.ah.shift(periods=-6)
```

```python
df.head(8)
```

### Limitation of the ML_DataFrame

```python
# drop related cols to target that are unknown

#df.drop(columns=['ah', 'rh'], inplace=True)
df.drop(columns=['rh'], inplace=True)
```

```python
# drop rows where target is unknown
df.dropna(subset=[target], inplace=True)
```

```python
df = df[features + [target]]
df
```

```python
# Missing value treatment
df = df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.shape
```

# Preparation Data

```python
from sklearn.model_selection import train_test_split
#training, test = train_test_split(df, test_size=0.2, shuffle=False, random_state=1999)
training, test = train_test_split(df, test_size=1000, shuffle=False, random_state=1999)
```

```python
X_train = training[features]
y_train = training[target]

x_test = test[features]
y_test = test[target]
```

```python
X_train
```

# Models/Classification

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
clf_reg = LinearRegression()
clf_reg.fit(X_train, y_train)
```

```python
pred_y = clf_reg.predict(x_test)
```

```python
df_test_pred = test[features]
df_test_pred["ah_target"] = test[target]
df_test_pred["pred"] = pred_y
df_test_pred
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred_y)
```
